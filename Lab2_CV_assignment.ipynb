```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Load the two images
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpeg')

# Resize images to the same dimensions if necessary
# image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Create a custom mask (you can design your own mask)
mask = np.zeros_like(image1)

# For example, create a linear gradient mask
mask[:, :mask.shape[1]//2] = (255, 255, 255)  # White on left side
mask[:, mask.shape[1]//2:] = (0, 0, 0)        # Black on right side

# Blend the images using the mask
blended_image = np.where(mask != 0, image1, image2)

# Display the blended image
cv2_imshow(blended_image)
```
